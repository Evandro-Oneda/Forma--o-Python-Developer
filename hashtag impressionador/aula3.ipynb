```python
import pandas as pd

tabela = pd.read_csv("C:\Formação Python Developer\Forma--o-Python-Developer\hashtag impressionador\clientes.csv")
#display(tabela)
display(tabela.info())

```

```python
# converter tabelas de texto para numeros

#display(tabela)

import sklearn
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
#display(tabela["profissao"])
codificador2 = LabelEncoder()
tabela["mix_credito"] = codificador2.fit_transform(tabela["mix_credito"])
codificador3 = LabelEncoder()
tabela["comportamento_pagamento"] = codificador3.fit_transform(tabela["comportamento_pagamento"])
display(tabela.info())

```

```python
# y é quem eu quero preve
y = tabela["score_credito"] 
 # x é quem a IA vai usar para prever entao tiramos a tabela de estudo e tabela inuteis
x = tabela.drop(columns=["score_credito", "id_cliente"])


# funcao para dividir aleatoriamente para treino e para teste
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)



```

```python
#criar um modelo de IA


from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# criando IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#treinando IA

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

# verificar qual melhor ia (acuracia)

previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste,previsao_arvoredecisao ))
display(accuracy_score(y_teste, previsao_knn))


```
